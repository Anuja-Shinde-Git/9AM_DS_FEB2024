# SUBPLOTS

```python
x = np.arange(1,10,.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x =  np.arange(1,10,.1)
y = np.tan(x)
plt.plot(x,y)
plt.show()
```

```python
# combining both the plot

x = np.arange(1,10,.1)
y1 = np.sin(x)
plt.plot(x,y1)

y2 = np.tan(x)
plt.plot(x,y2)
plt.show()
```

```python
# rearranging the code for better readability !!!

x = np.arange(1,10,.1)
y1 = np.sin(x)
y2 = np.tan(x)

plt.plot(x,y1)
plt.plot(x,y2)

plt.show()
```

```python
# using the subplots


# subplots >> is nothing but dividing canvas into small parts
# here we use packinga and unpacking concept of tuple
# method >> fig,ax = plt.subplots(1,3)
# fig >> figure 
# ax >> axis (no of axis)
```

```python
x = np.arange(1,10,.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

fig,ax = plt.subplots(1,3)
ax[0].plot(x,y1)
ax[1].plot(x,y2)
ax[2].plot(x,y3)
```
