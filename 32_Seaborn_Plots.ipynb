```python
'''
Definitions
Median
The median (middle quartile) marks the mid-point of the data and is shown by the line that divides the box into two parts. 
Half the scores are greater than or equal to this value and half are less.

Inter-quartile range
The middle “box” represents the middle 50% of scores for the group. 
The range of scores from lower to upper quartile is referred to as the inter-quartile range. 
The middle 50% of scores fall within the inter-quartile range.

Upper quartile
Seventy-five percent of the scores fall below the upper quartile.

Lower quartile
Twenty-five percent of scores fall below the lower quartile.

Whiskers
The upper and lower whiskers represent scores outside the middle 50%. 
Whiskers often (but not always) stretch over a wider range of scores than the middle quartile groups.

'''
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.boxplot(data=iris,x='species',y='sepal_length')
```

```python
# lower right corner small dot is called as "outlire" >> very different value from all of the values so that we cant consider
```

# Correlation

```python
# using correlation we can see numeric values of data variation and relation between them 
# and from that we can analyis data
# the values near about 1 is more related and diagonal one's are already 1 we cant consider them as they are of same components values
```

```python
iris
```

```python
# iris.corr()
# ValueError: could not convert string to float: 'setosa'
# # only for numerical values  so that we slice the data using iloc and take only 1st four columns and the calculate correlation
```

```python
ans = iris.iloc[:,0:4]
```

```python
ans
```

```python
# using correlation we can see numeric values of data variation and relation between them 
# and from that we can analyis data
# the values near about 1 is more related and diagonal one's are already 1 we cant consider them as they are of components values

ans.corr()
```
