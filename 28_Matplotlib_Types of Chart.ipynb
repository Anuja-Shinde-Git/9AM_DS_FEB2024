# 1.BAR

```python
# to use bar graph use method plt.bar
```

```python
# we have fruit shop and and no.of fruits in it we are going to understand that info through bar graph

# shop 1
# fruit 1 : 10
# fruit 2 : 20
# fruit 3 : 30
```

```python
plt.bar([1,2,3],[10,20,5])
plt.show
```

```python
# here width of bar is to large we can reduce it using width fuction and adding legend

plt.bar([1,2,3],[10,20,5],width=0.5)
plt.legend('shop1')
plt.show()
```

```python
# shop 2
# fruit 1 : 5
# fruit 2 : 10
# fruit 3 : 6
```

```python
plt.bar([1,2,3],[5,10,6],width = 0.5)
plt.legend('shop2')
plt.show()
```

```python
# plotting both the graphs in single canvas
```

```python
plt.bar([1,2,3],[10,20,5],width = 0.5)
plt.bar([1,2,3],[5,10,6],width = 0.5)
plt.legend(['shop1','shop2'])
plt.show()
```

```python
# not understanding the readings so we keep default width for shop1 and 0.5 width for shop2 for better understanding

plt.bar([1,2,3],[10,20,5])
plt.bar([1,2,3],[5,10,6],width = 0.5)
plt.legend(['shop1','shop2'])
plt.show()

```

```python
# making both graph side by side >> 

plt.bar([1,2,3],[10,20,5])
plt.bar([1.5,2.5,3.5],[5,10,6],width=0.5)
plt.legend(['shop1','shop2'])
plt.show()
```

```python
# adjusting width for better reading

plt.bar([1,2,3],[10,20,5],width=0.5)
plt.bar([1.5,2.5,3.5],[5,10,6],width=0.5)
plt.legend(['shop1','shop2'])
plt.show()
```

```python
# adjusting width for better reading


plt.bar([1,2,3],[10,20,5],width=0.3)
plt.bar([1.5,2.5,3.5],[5,10,6],width=0.3)
plt.legend(['shop1','shop2'])
plt.show()
```

```python
# making side by side

plt.bar([1,2,3],[10,20,5],width=0.3)
plt.bar([1.3,2.3,3.3],[5,10,6],width=0.3)
plt.legend(['shop1','shop2'])
plt.show()
```

# HISTOGRAM

```python
# histogram is used to show categorical data / group data
# understand with the help of below example

# MARKS : 0,1,2,3,4,5......89,97,98,99,100

# GRADE : 
#        0-35 >> F
#        36-60 >> C
#        61-75 >> B
#        76-100 >> A

# for grade system we can use histogram plot
```

```python
# for checking distribution
```

```python
vals = [1,2,3,1,2,3,3,2,3,2,2,2]
plt.hist(vals)
# plt.show()

# showing no.of times for each value
# 1 >> 2 times
# 2 >> 6 times
# 3 >> 4 times

# below two arrays displying info
# array 1 >> showing frequency of array 2's values
# array 2 >> showing values and distributed given values from upper limit and lower limit in 10 parts called artists

# artists >> bin size
#  bin size is noting but no.of groups on which you want to distributes the values
```

```python
vals = [1,2,3,1,2,3,3,2,3,2,2,2,1.5]
plt.hist(vals)
# plt.show()
```

```python
# we can change the default bin size
#  bin size is noting but no.of groups on which you want to distributes the values

vals = [1,2,3,1,2,3,3,2,3,2,2,2]
plt.hist(vals)
# plt.show()
```

```python
vals = [1,2,3,1,2,3,3,2,3,2,2,2]
plt.hist(vals,bins=5)
# plt.show()
```

```python
vals = [1,2,3,1,2,3,3,2,3,2,2,2]
plt.hist(vals,bins=3)
# plt.show()
```

```python
vals = [1,2,3,1,2,3,3,2,3,2,2,2]
plt.hist(vals,bins=2)
# plt.show()

# bins = 2 distributes values in 2 part 1 to 2 and 2 to 3
```
