```python
# generating custom index >> generating index as per user concern
# best way to generate list is using list comprehension
# want to generate square of index
for i in range(11):
    print(i**2)
```

```python
[i**2 for i in range(11)]
```

```python
# here we create custom index
```

```python
df['custom'] = [i**2 for i in range(1,1705)]
# here we take 1705 because list consider end index as n-1
```

```python
df
```

```python
# now we want to replace original index's place by custom index
```

```python
# to show original index
df.index
```

```python
# now chaange orin=ginal index by custom
df.index = df['custom']
```

```python
df
```

```python
# now remove right side custom index
df.drop('custom',axis = 1,inplace = True)
# here two custom col are present still exactly right side custom will removed 
# as original index is never deleted 
```
