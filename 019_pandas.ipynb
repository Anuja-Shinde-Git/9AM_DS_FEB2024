```python
result.merge(students,left_on = 'seating no',right_on = 'Roll no',how = 'inner')
# how = inner is default value
# acts as inner join return matched values >> expected result
```

```python
# try with different values for how=?
```

```python
# how = outer
result.merge(students, left_on = 'seating no', right_on = 'Roll no', how = 'outer')
# acts as outer join all matched and unmatched values return
```

```python
# how = right
result.merge(students,left_on='seating no',right_on='Roll no',how='right')
# same o/p as how = outer
```

```python
# how = left
result.merge(students,left_on='seating no',right_on='Roll no',how='left')
# same o/p as how = inner
```
