# Removing duplicates

```python
# finding duplicates in dataframe 
```

```python
# to find duplicate we have to created boolen mask 
df.duplicated()
# no duplicates hence all are false if there is duplicate return true
```

```python
# using boolean masking
# there is no true value means no duplicates in the dataset
df[df.duplicated()]
```

```python
df
```

```python
# here we created duplicate rows 
df.loc[1706] = pd.Series(d2)
df
```

```python
df.loc[1707] = pd.Series(d2)
df
```

```python
# here we check for duplicate values
df.duplicated()
```

```python
# here values of 3 row are same and showing 2 rows as True i.e. duplicate because panda consider one row as original
# and remaining two are duplicate so in sequence 1st row come consider that one is original here 1705 is original
```

```python
df[df.duplicated()]
# using masking
```

```python
# to remove duplicate use drop method
df.drop_duplicates()
# by default it will drop last two values 1706 and 1707
```

```python
df
# inplace not changed so all values are as it is
```

```python
# here we can decide which duplicate values we want to remove using keep last and first
# if  keep = last, it will remove first two i.e 1705 and 1706 ,
df.drop_duplicates(keep = 'last')
# will remove duplicate 1705 and 1706
```

```python
df
# not placed inplace changes so all values are as it is
```

```python
# df.drop_duplicates(keep = 'last',inplace = True) we run this inplce changes takes place and data permantaly removes
# careful with data while using inplace
```

```python
# drop is not noly used to remove duplicate we can remove any value using index
df.drop(2)
# removed index 2 but only temperary as inplce not used
```

```python
import pandas as pd

# Your existing DataFrame
df = pd.DataFrame({'A': [1, 2, 3],
                   'B': [4, 5, 6]})

# Dictionary representing the new row
d = {'A': 7, 'B': 8}

# Convert the dictionary to a Series
new_row_series = pd.Series(d)

# Add the new row to the DataFrame using loc
df.loc[len(df)] = new_row_series

print(df)
```
