# Some complex plots

```python
# Square function
x_val = [0,1,2,3]
y_val = [0,1,4,9]
plt.plot(x_val,y_val)
plt.show
```

```python
# for clear square function we increase no.of points
x_val = [-3,-2,-1,0,1,2,3]
y_val = [9,2,1,0,1,4,9]
plt.plot(x_val,y_val)
plt.show()
```

```python
# using numpy generating numbers 
x_val = [-3,-2,-1,0,1,2,3]
y_val = x_val**2
plt.plot(x_val,y_val)
plt.show()
```

```python
# for elementwise operation need to make x_val as array
x_val = np.array([-3,-2,-1,0,1,2,3])
y_val = x_val**2
plt.plot(x_val,y_val)
plt.show()
```

```python
# for more smooth curve for square function need to add more points >> using arange >> (start:end-1:step)
x_val = np.arange(-3,4,1)
y_val = x_val**2
plt.plot(x_val,y_val)
plt.show()
```

```python
# step size .1 generates more points

x_val = np.arange(-3,4,.1)
y_val = x_val**2
plt.plot(x_val,y_val)
plt.show()
```

```python
# for symetri values

x_val = np.arange(-4,4,.1)
y_val = x_val**2
plt.plot(x_val,y_val)
plt.show()
```

```python
# xticks = .1 shows very small values so use little greater value

x_val = np.arange(-4,4,.1)
y_val = x_val**2
plt.xticks(np.arange(-4,4,.3))
plt.plot(x_val,y_val)
plt.show()
```

```python
# still not clear view so we use parameter rotation 90

x_val = np.arange(-4,4,.1)
y_val = x_val**2
plt.xticks(np.arange(-4,4,.3),rotation = 90)
plt.plot(x_val,y_val)
plt.show()
```
