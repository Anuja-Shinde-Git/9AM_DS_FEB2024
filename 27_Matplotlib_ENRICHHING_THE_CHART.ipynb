# adding legends

```python
# specifing the graph for more identification called adding legends ie. which graph belongs to which function (notations)
# adding lable legend while plotting
```

```python
x_val = np.arange(-4,4,.1)
plt.xticks(np.arange(-4,4,.3),rotation=90)

y1_val = x_val**2
y2_val = x_val**3

plt.plot(x_val,y1_val,label = 'square')
plt.plot(x_val,y2_val,label = 'cube')

plt.legend()
plt.show()
```

```python
# we can also add label in legend also

x_val = np.arange(-4,4,.1)
plt.xticks(np.arange(-4,4,.3),rotation=90)

y1_val = x_val**2
y2_val = x_val**3

plt.plot(x_val,y1_val)
plt.plot(x_val,y2_val)

plt.legend(['square','cube'])
plt.show()
```

```python
# changing the default size of label to the lower right corner using location in legend (loc)

x_val = np.arange(-4,4,.1)
plt.xticks(np.arange(-4,4,.3),rotation=90)

y1_val = x_val**2
y2_val = x_val**3

plt.plot(x_val,y1_val)
plt.plot(x_val,y2_val)

plt.legend(['square','cube'],loc ='lower right')
plt.show()
```

```python
# removing the lebels frame/labelbox using frameon = False

x_val = np.arange(-4,4,.1)
plt.xticks(np.arange(-4,4,.3),rotation=90)

y1_val = x_val**2
y2_val = x_val**3

plt.plot(x_val,y1_val)
plt.plot(x_val,y2_val)

plt.legend(['square','cube'], loc ='lower right',frameon = False)
plt.show()
```

```python
# changing the label colummn into 2 column using no. of column (ncol=2)

x_val = np.arange(-4,4,.1)
plt.xticks(np.arange(-4,4,.3),rotation=90)

y1_val = x_val**2
y2_val = x_val**3

plt.plot(x_val,y1_val)
plt.plot(x_val,y2_val)

plt.legend(['square','cube'],loc='lower right',frameon=False,ncol=2)
plt.show()
```

```python
# pointing some points on graph >> need to give positional arguments x and y points
# we want to add some text on intersection points using plt.text(x,y,'text')
# so here only intersection points 0 and 1 as square and cube of 0 and 1 are always 0 and 1
# also we can zoom in the gragh to see zoom in view using limit function ie. xlim (x,y) ylim(x,y)
# all the required function should run befor show() function


x_val = np.arange(-4,4,.1)
plt.xticks(np.arange(-4,4,.3),rotation=90)

y1_val = x_val**2
y2_val = x_val**3

plt.plot(x_val,y1_val)
plt.plot(x_val,y2_val)

plt.legend(['square','cube'],loc='lower right',frameon=False,ncol=2)

# adding text
plt.text(0,0,'(0,0)')
plt.text(1,1,'(1,1')

# zooming in
plt.xlim(-2,2)
plt.ylim(-2,2)

plt.show()
```
